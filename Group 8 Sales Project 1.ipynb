```python
# Function to check if Promo2 is active for each row
def is_promo2_active(row):
    if row['Promo2'] == 1:
        if (row['Date'].year > row['Promo2SinceYear']) or \
           (row['Date'].year == row['Promo2SinceYear'] and row['Date'].isocalendar().week >= row['Promo2SinceWeek']):
            if row['PromoInterval'] != 'None':
                # Get the month abbreviation (e.g., 'Jan', 'Feb', etc.)
                month = row['Date'].strftime('%b')
                if month in row['PromoInterval'].split(','):
                    return 1
    return 0
```

```python
# Apply the function to the merged_df
merged_df['Promo2Active'] = merged_df.apply(is_promo2_active, axis=1)
```
